```python
import matplotlib.pyplot as plt
```

# Line Plot :- 

---


Give the parameter and plot a simple line

plt.plot([]) :- give parameter x-axis and y-axis

plt.show() :- use to display the plot.

You can give title name using  **plt.title('')**

```python
plt.plot([1,2,3])
plt.title('Information')
plt.show()
```

You can also give the labels name of x-axis and y-axis using **plt.xlabel() and plt.ylabel()** 

```python
plt.plot([1,2,3])
plt.title('Information')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Bar Plot

---



In this plot, we use NumPy library as np. It is used to compare the data for this we can create horizontal and vertical bars.

**plt.bar()** :- take parameter for plotting. 

```python
import numpy as np
x = [30,40,50,60,20]
y = [1,3,5,6,7]

plt.bar(x,y)
plt.title("Bar Plot")
plt.xlabel("Salary")
plt.ylabel("Employee")
plt.show()
```

# Pie plot or Chart

---



Create a chart of four section 

using **plt.pie(parameters)** plot a circular chart with equal axis.

```python
labels = ['Brownie','Choclate','Ice-cream','Cakes']
count = [20,30,10,5]
colors = ['brown','green','pink','red']

plt.pie(count, labels=labels,colors=colors)

plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

# Histogram 

---



It shows the distribution of data.

plt.hist() use to show distribution among data.

```python
age_group = [5,15,25,35]
age = [5,10,15,20,25,30]

plt.hist(age_group,bins=age,histtype='bar')
plt.show()
```

#Scatter Plot

---



It is plotting certain points and then comparing between points, only show points using **plt.scatter()**

```python
x = [4,8,7,5,3,8,1]
y = [3,7,9,4,8,2,6]

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Plot A Numpy Array

---

Create an nd array on x-axis and store equation value on y-axis.


```python
x_array = np.arange(3,21)
y_equation = 2 * x_array + 3

plt.title("Numpy array Plot")

plt.plot(x_array,y_equation)
plt.show()
```
